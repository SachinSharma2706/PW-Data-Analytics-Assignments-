## 1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make inferences about a population based on a sample. It involves testing an assumption (hypothesis) about a population parameter and determining whether the evidence from the sample data supports or rejects that assumption.




## 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- **Null Hypothesis (H₀):** Assumes no effect or no difference in the population (default assumption).
- **Alternative Hypothesis (H₁ or Ha):** Suggests there is an effect or a difference.
- Hypothesis testing aims to determine if there is enough evidence to reject the null hypothesis.


## 3. What is the significance level in hypothesis testing, and why is it important?
- The **significance level (α)** is the probability threshold for rejecting the null hypothesis.
- Common values: **0.05 (5%)** or **0.01 (1%)**.
- A lower α reduces the chance of a **Type 1 error (false positive)**.


## 4. What does a P-value represent in hypothesis testing?
- The **P-value** measures the probability of obtaining test results at least as extreme as the observed data, assuming the null hypothesis is true.
- A **small P-value (< α)** suggests strong evidence against H₀, leading to its rejection.



## 5. How do you interpret the P-value in hypothesis testing?
- **P < α (e.g., 0.05):** Reject H₀ (significant result).
- **P ≥ α:** Fail to reject H₀ (insufficient evidence).

## 6. What are Type 1 and Type 2 errors in hypothesis testing?
- **Type 1 Error (False Positive):** Rejecting a true null hypothesis (incorrectly claiming an effect).
- **Type 2 Error (False Negative):** Failing to reject a false null hypothesis (missing a real effect).


## 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- **One-tailed test:** Tests for an effect in **one direction** (e.g., greater than or less than).
- **Two-tailed test:** Tests for an effect in **both directions** (e.g., significantly different).


## 8. What is the Z-test, and when is it used in hypothesis testing?
- A **Z-test** is used when comparing sample data to a known population with a **large sample size (n ≥ 30)** and a known population standard deviation.
- Commonly used for **proportions and means**.

## 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- **Z-score formula:**
  \[
  Z = \frac{X - \mu}{\sigma}
  \]
  Where:
  - \(X\) = sample mean
  - \(\mu\) = population mean
  - \(\sigma\) = population standard deviation
- The **Z-score** indicates how many standard deviations a data point is from the population mean.

## 10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The **T-distribution** is used when the **sample size is small (n < 30)** and the **population standard deviation is unknown**.
- It is similar to the normal distribution but has **fatter tails**, accounting for more variability in small samples.

## 11. What is the difference between a Z-test and a T-test?
- **Z-test** is used when:
  - Sample size is **large (n ≥ 30)**.
  - Population standard deviation (\(\sigma\)) is **known**.
  - Data follows a **normal distribution**.
- **T-test** is used when:
  - Sample size is **small (n < 30)**.
  - Population standard deviation (\(\sigma\)) is **unknown**.
  - Uses **T-distribution** instead of the normal distribution.

## 12. What is the T-test, and how is it used in hypothesis testing?
- The **T-test** is used to compare means between groups when the population standard deviation is unknown.
- Types:
  - **One-sample T-test**: Compares a sample mean to a known population mean.
  - **Two-sample T-test**: Compares means of two independent groups.
  - **Paired T-test**: Compares means from the same group before and after an intervention.


## 13. What is the relationship between Z-test and T-test in hypothesis testing?
- Both **Z-tests and T-tests** test hypotheses about **population means**.
- **T-tests** are used when the **sample size is small** or **σ is unknown**.
- **Z-tests** are preferred for **large samples** or when **σ is known**.


## 14. What is a confidence interval, and how is it used to interpret statistical results?
- A **confidence interval (CI)** gives a range of values within which the true population parameter is likely to lie.
- **Formula**:
  \[
  CI = \bar{X} \pm (Z_{\alpha/2} \times SE)
  \]
  Where:
  - \(\bar{X}\) = sample mean,
  - \(Z_{\alpha/2}\) = critical value (from Z-table or T-table),
  - \(SE\) = standard error.
- A **95% confidence interval** means that if we repeated the sampling many times, 95% of the intervals would contain the true mean.


## 15. What is the margin of error, and how does it affect the confidence interval?
- **Margin of error (MOE)** is the amount of uncertainty around the sample estimate.
- **Formula**:
  \[
  MOE = Z_{\alpha/2} \times SE
  \]
- A **larger margin of error** means a **wider confidence interval**, indicating less precision.
- **Reducing MOE** requires a **larger sample size**.


## 16. How is Bayes' Theorem used in statistics, and what is its significance?
- **Bayes' Theorem** helps update probabilities based on new evidence.
- **Formula**:
  \[
  P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
  \]
  Where:
  - \(P(A|B)\) = Probability of event A given B,
  - \(P(B|A)\) = Probability of event B given A,
  - \(P(A)\) = Prior probability of A,
  - \(P(B)\) = Prior probability of B.
- Used in **medical testing, spam filtering, and machine learning**.


## 17. What is the Chi-square distribution, and when is it used?
- The **Chi-square distribution** is used for tests involving **categorical data**.
- Properties:
  - **Right-skewed** and only **positive values**.
  - Used when **comparing observed and expected frequencies**.
- Used in:
  - **Chi-square goodness-of-fit test**.
  - **Chi-square test for independence**.

## 18. What is the Chi-square goodness of fit test, and how is it applied?
- **Purpose**: Tests if an observed frequency distribution matches an expected distribution.
- **Formula**:
  \[
  \chi^2 = \sum \frac{(O - E)^2}{E}
  \]
  Where:
  - \(O\) = observed frequency,
  - \(E\) = expected frequency.
- **Example**: Testing if dice rolls are fair.


## 19. What is the F-distribution, and when is it used in hypothesis testing?
- The **F-distribution** is used to compare **variances** between two groups.
- Properties:
  - **Right-skewed** and depends on **degrees of freedom**.
- Used in:
  - **ANOVA tests**.
  - **Regression analysis**.


## 10. What is an ANOVA test, and what are its assumptions?
- **Analysis of Variance (ANOVA)** is used to compare **means of three or more groups**.
- **Null hypothesis (H₀)**: All group means are equal.
- **Alternative hypothesis (H₁)**: At least one group mean is different.
- **Assumptions**:
  1. **Independence**: Samples are independent.
  2. **Normality**: Data in each group follows a normal distribution.
  3. **Homogeneity of variance**: Variances are equal across groups.


## 21. What are the different types of ANOVA tests?
ANOVA (**Analysis of Variance**) is used to compare means across multiple groups. The main types include:

### **a) One-Way ANOVA**
- Compares means of **three or more** independent groups based on **one factor**.
- **Example:** Comparing test scores across three different teaching methods.

### **b) Two-Way ANOVA**
- Compares means when there are **two independent factors**.
- **Example:** Studying how both **teaching method** and **gender** affect test scores.

### **c) Repeated Measures ANOVA**
- Used when the **same subjects** are tested under **different conditions**.
- **Example:** Measuring blood pressure **before, during, and after** taking a drug.

### **d) MANOVA (Multivariate Analysis of Variance)**
- Extends ANOVA to compare **multiple dependent variables**.
- **Example:** Evaluating the effect of diet on **both blood pressure and cholesterol levels**.


## 22. What is the F-test, and how does it relate to hypothesis testing?
- The **F-test** is used to compare **variances** or **multiple means** in hypothesis testing.
- It follows the **F-distribution**, which is right-skewed and depends on degrees of freedom.

### **F-test Formula (for ANOVA)**
\[
F = \frac{\text{Between-group variance}}{\text{Within-group variance}}
\]
Where:
- **A high F-value** suggests **a significant difference** between groups.
- **A low F-value** suggests **no significant difference**.

### **Uses of the F-test**
1. **In ANOVA:** Compares means of multiple groups.
2. **In Regression Analysis:** Checks the overall significance of a model.
3. **In Variance Testing:** Compares two population variances.

# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
# Practical Hypothesis Testing with Python

#


import scipy.stats as stats

# Given population parameters
mu = 50  # Population mean
sigma = 10  # Population standard deviation
n = 30  # Sample size

# Sample data
sample_mean = 53

# Calculate Z-score
z_score = (sample_mean - mu) / (sigma / (n ** 0.5))

# Calculate P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis (H₀). The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H₀). No significant difference detected.")
```

# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
def plot_type_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Standard Normal Distribution")

    # Type 1 Error (False Positive)
    z_alpha = stats.norm.ppf(1 - alpha)
    plt.fill_between(x, y, where=(x >= z_alpha), color='red', alpha=0.3, label="Type 1 Error (α)")

    # Type 2 Error (False Negative)
    z_beta = stats.norm.ppf(beta)
    plt.fill_between(x, y, where=(x <= z_beta), color='blue', alpha=0.3, label="Type 2 Error (β)")

    plt.axvline(z_alpha, color='black', linestyle='dashed', label=f"Critical Value (Zα): {z_alpha:.2f}")
    plt.axvline(z_beta, color='purple', linestyle='dashed', label=f"Power (1 - β)")

    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

plot_type_errors()
```

## 6. Perform an Independent T-Test and Interpret the Results

```python

import numpy as np
import scipy.stats as stats

# Generate random sample data for two independent groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)

# Perform independent T-test
t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis (H₀). The two groups have significantly different means.")
else:
    print("Fail to reject the null hypothesis (H₀). No significant difference between the groups.")
```

# 7. Perform a Paired Sample T-Test and Visualize the Comparison Results

```python
import matplotlib.pyplot as plt

# Generate paired sample data (before and after)
np.random.seed(42)
before = np.random.normal(loc=50, scale=10, size=30)
after = before + np.random.normal(loc=2, scale=5, size=30)  # Small increase after treatment

# Perform paired T-test
t_stat, p_value = stats.ttest_rel(before, after)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis (H₀). The two conditions are significantly different.")
else:
    print("Fail to reject the null hypothesis (H₀). No significant change between conditions.")

# Visualization
plt.figure(figsize=(8, 5))
plt.scatter(before, after, color="blue", alpha=0.6, label="Before vs After")
plt.plot([40, 70], [40, 70], 'r--', label="No Change Line")  # Reference line
plt.xlabel("Before Treatment")
plt.ylabel("After Treatment")
plt.title("Paired Sample T-Test: Before vs After")
plt.legend()
plt.show()
```

# 9. Calculate the Confidence Interval for a Sample Mean and Explain Its Significance

```python
def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)

    # Calculate margin of error
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
ci_lower, ci_upper = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)

    # Critical value for t-distribution
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_critical * (sample_std / np.sqrt(n))

    return margin

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
moe = margin_of_error(sample_data)
print(f"Margin of Error (95% confidence): {moe:.2f}")
```

# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the processD

```python
def bayes_theorem(prior, likelihood, marginal):
    posterior = (likelihood * prior) / marginal
    return posterior

# Example: Disease testing
prior_prob = 0.01  # Prior probability of having a disease
likelihood = 0.95  # Probability of testing positive given the disease
false_positive = 0.05  # Probability of testing positive without the disease

# Total probability of testing positive
marginal_prob = (likelihood * prior_prob) + (false_positive * (1 - prior_prob))

# Applying Bayes' Theorem
posterior_prob = bayes_theorem(prior_prob, likelihood, marginal_prob)
print(f"Updated probability of having the disease given a positive test: {posterior_prob:.4f}")
```

# 18. Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Generate sample data
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)

# Perform One-Way ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)

# Print results
print(f"ANOVA F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print("Conclusion:", "Reject H₀: Significant difference between groups" if p_value < 0.05 else "Fail to reject H₀: No significant difference")

# Prepare data for visualization
df = pd.DataFrame({"Values": np.concatenate([group1, group2, group3]),
                   "Group": ["Group 1"]*30 + ["Group 2"]*30 + ["Group 3"]*30})

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x="Group", y="Values", data=df, palette="Set2")
plt.title("One-Way ANOVA: Group Comparisons")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```

# 19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

def check_anova_assumptions(*groups):
    print("### Normality Test (Shapiro-Wilk) ###")
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"Group {i+1}: W={stat:.3f}, p={p:.4f} -> {'Normal' if p > 0.05 else 'Not Normal'}")

    stat, p = stats.levene(*groups)
    print("\n### Homogeneity of Variance (Levene's Test) ###")
    print(f"Levene's Test: W={stat:.3f}, p={p:.4f} -> {'Equal Variance' if p > 0.05 else 'Unequal Variance'}")

    df = pd.DataFrame({"Values": np.concatenate(groups),
                       "Group": sum([[f"Group {i+1}"]*len(g) for i, g in enumerate(groups)], [])})

    plt.figure(figsize=(8, 6))
    sns.boxplot(x="Group", y="Values", data=df, palette="Set2")
    plt.title("Group Distributions")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.show()

np.random.seed(42)
g1 = np.random.normal(50, 10, 30)
g2 = np.random.normal(55, 10, 30)
g3 = np.random.normal(60, 10, 30)

check_anova_assumptions(g1, g2, g3)
```

# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

np.random.seed(42)
size = 30
factor_A = np.repeat(["A1", "A2"], size)
factor_B = np.tile(["B1", "B2"], size)
values = np.random.normal(50, 10, 2*size) + (np.array(factor_A) == "A2") * 5 + (np.array(factor_B) == "B2") * 3

df = pd.DataFrame({"Factor_A": factor_A, "Factor_B": factor_B, "Values": values})

model = ols("Values ~ C(Factor_A) * C(Factor_B)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

plt.figure(figsize=(8, 6))
sns.boxplot(x="Factor_A", y="Values", hue="Factor_B", data=df, palette="Set2")
plt.title("Two-Way ANOVA: Interaction Between Factors")
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=10, size=30)

t_stat, p_value = stats.ttest_ind(sample1, sample2)

print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
print("Conclusion:", "Reject H₀: Means are significantly different" if p_value < 0.05 else "Fail to reject H₀: No significant difference")
```

#24 . Perform a hypothesis test for population variance using a Chi-square distribution and interpret the resultsD

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)
sample_var = np.var(sample, ddof=1)
pop_var = 100
df = len(sample) - 1
chi_square_stat = (df * sample_var) / pop_var
p_value = 1 - stats.chi2.cdf(chi_square_stat, df)

print(f"Chi-Square Statistic: {chi_square_stat:.2f}, P-value: {p_value:.4f}")
print("Conclusion:", "Reject H₀: Population variance is different" if p_value < 0.05 else "Fail to reject H₀: No significant difference")
```

#25 .Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
import scipy.stats as stats

# Sample data
group1_successes = 120
group1_total = 200
group2_successes = 150
group2_total = 250

# Proportions
p1 = group1_successes / group1_total
p2 = group2_successes / group2_total

# Pooled proportion
p_pool = (group1_successes + group2_successes) / (group1_total + group2_total)

# Standard error
se = np.sqrt(p_pool * (1 - p_pool) * (1 / group1_total + 1 / group2_total))

# Z-statistic
z_stat = (p1 - p2) / se

# P-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat:.2f}, P-value: {p_value:.4f}")
print("Conclusion:", "Reject H₀: Proportions are significantly different" if p_value < 0.05 else "Fail to reject H₀: No significant difference")
```

# 26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Sample data
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 15, 30)

# Variances
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

# F-statistic
f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(group1) - 1
df2 = len(group2) - 1

# P-value from the F-distribution (right-tailed test)
p_value = 1 - stats.f.cdf(f_stat, df1, df2)

# Results
print(f"F-statistic: {f_stat:.2f}, P-value: {p_value:.4f}")
print("Conclusion:", "Reject H₀: Significant difference in variances" if p_value < 0.05 else "Fail to reject H₀: No significant difference")

# Visualization of F-distribution and F-statistic
x = np.linspace(0, 5, 1000)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f"F-distribution (df1={df1}, df2={df2})", color='blue')
plt.fill_between(x, y, where=(x >= f_stat), color='red', alpha=0.3, label=f"F-statistic region (F={f_stat:.2f})")
plt.axvline(f_stat, color='red', linestyle='dashed')
plt.title("F-Test: Comparison of Variances")
plt.xlabel("F-value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

# 27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats

# Simulate observed data (frequencies)
observed = np.array([30, 50, 20])  # Example observed frequencies for 3 categories

# Expected frequencies (uniform distribution, assuming equal expected frequencies)
total = np.sum(observed)
expected = np.array([total / len(observed)] * len(observed))

# Perform Chi-square test for goodness of fit
chi2_stat, p_value = stats.chisquare(observed, expected)

# Results
print(f"Chi-square Statistic: {chi2_stat:.2f}, P-value: {p_value:.4f}")
print("Conclusion:", "Reject H₀: Data does not fit the expected distribution" if p_value < 0.05 else "Fail to reject H₀: Data fits the expected distribution")
```
